```python
df = pd.read_csv('Data/superhero_info - superhero_info.csv')
df.head()
```

# Separating Hero and Publisher Column

```python
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
df = df.drop(columns=['Hero|Publisher'])
df.head()
```

# Replacing all single quotes in Measurements with double quotes

```python
## use .str.replace to replace all single quotes
df['Measurements'] = df['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements'].head()
```

```python
#Ensuring it worked
type(df.loc[0,'Measurements'])
```

```python
#convert the single "Measurements" column into 2 separate columns
weight_height = df['Measurements'].apply(pd.Series)
weight_height
```

```python
weight_height['Weight'] = weight_height['Weight'].str.split(' ',expand=True)[0].astype('float32')
weight_height
```

```python
weight_height['Height'] = weight_height['Height'].str.split(' ',expand=True)[0].astype('float32')
weight_height
```

# Concatenate Weight_Height column to Dataframe

```python
df = pd.concat((df, weight_height), axis = 1)
df.head()
```

```python
#Dropping Measurements column
df = df.drop(columns = 'Measurements')
df.head()
```

# Adding One Hot Encoded Powers to Original Dataframe

```python
#Importing new Powers Data
power = pd.read_csv("Data/superhero_powers - superhero_powers.csv")
power.head()
```

```python
#Creating a list
power['Power_List'] = power['Powers'].str.split(',')
power.head()
```

```python
power_exploded = power.explode('Power_List')[['hero_names', 'Power_List']]
power_exploded
```

```python
cols_to_make = power_exploded['Power_List'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    power[col] = power['Powers'].str.contains(col)
power.head()
```

```python
power = power.drop(columns = ['Powers', 'Power_List'])
power = power.rename(columns = {'hero_names': 'Hero'})
power = power.rename(columns = {'hero_names': 'Hero'})
power.head()
```

# Creating Final Dataframe

```python
df = pd.merge(df,power, on='Hero')
df.head()
```
